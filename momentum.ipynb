```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def one_step(x0, nb_iter):
    x = [x0]
    for _ in range(nb_iter):
        x.append(x[-1]*0.01 + 0.99)
    return x
```

```python
x = one_step(5, 100)
plt.plot(x)
```

```python
def two_step(x0, x1, nb_iter, a,b):
    x = [x0, x1]
    for _ in range(nb_iter):
        x.append(x[-1]*a + b*x[-2])
    return x
```

```python
y = two_step(11, 12, 20, 1/2, 1/2)
plt.plot(y, label="2 step")
plt.legend()
```

```python
def P(x):
    return 2*x-2

def iterative_algo(x0, alpha, nb_iter=50):
    ls = [P(x0)]
    for _ in range(nb_iter):
        ls.append(P(ls[-1] - alpha*P(ls[-1])))
    return ls
```

```python
x = iterative_algo(10, 0.005)
plt.plot(x)
```

```python
x = iterative_algo(10, 0.005)
plt.plot(x)
```

```python
def beta(L):
    alpha = [1/l for l in L]
    return [(-a + np.sqrt(a**2+2))/2 for a in alpha]

L = [0.2, 0.5, 0.7, 1, 2, 3, 4, 5, 7, 10]
plt.plot(beta(L))
```
